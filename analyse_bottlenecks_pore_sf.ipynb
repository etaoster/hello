```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

def read_bottleneck_data(csv_path):    
    
    ### Loop the data lines
    with open(csv_path, 'r') as temp_f:
    # get No of columns in each line
        col_dict = [ len(l.split(",")) for l in temp_f.readlines() ]

    ### Generate column names  (names will be 0, 1, 2, ..., maximum columns - 1)
    column_names = [i for i in range(0, max(col_dict))]

    ### Read csv
    df = pd.read_csv(csv_path, header=None, delimiter=",", names=column_names, skiprows=2)
    df = df.rename(columns = {0:'Snapshot', 1:'Tunnel cluster', 2:'Tunnel', 3:'Throughput', 4:'Cost', 5:'x', 6:'y', 7:'z', 8:'Bottleneck radius', 9:'br', 10:'br', 11:'br',12:'br', 13:'br',14:'br',15:'br', 16:'br',17:'br',18:'br',19:'br',20:'br',21:'br',22:'br',23:'br'})

    return df

def extract_bradii(df, tun_cl):
    """Return a filtered df of 'Bottleneck radius' for a given tunnel cluster"""
    return df[df['Tunnel cluster'] == tun_cl][['Bottleneck radius']]
```

```python
df_6agf_1 = read_bottleneck_data("/home/elaine/elaine_data/caver_3.0/pdbs_0_250/6AGF_WTR_rep_1/out/analysis/bottlenecks.csv")
df_6agf_2 = read_bottleneck_data("/home/elaine/elaine_data/caver_3.0/pdbs_0_250/6AGF_WTR_rep_2/out/analysis/bottlenecks.csv")
df_6agf_3 = read_bottleneck_data("/home/elaine/elaine_data/caver_3.0/pdbs_0_250/6AGF_WTR_rep_3/out/analysis/bottlenecks.csv")

df_6agf_1_pore = extract_bradii(df_6agf_1, 3)
df_6agf_1_sf = extract_bradii(df_6agf_1, 4)

df_6agf_2_pore = extract_bradii(df_6agf_2, 5)
df_6agf_2_sf = extract_bradii(df_6agf_2, 4)

df_6agf_3_pore = extract_bradii(df_6agf_3, 5)
df_6agf_3_sf = extract_bradii(df_6agf_3, 4)


fig, axes = plt.subplots(1, 2, figsize=(20, 10), sharey=True)
fig.suptitle('6AGF (Nav1.4)')

sns.distplot(df_6agf_1_pore, ax=axes[0], label='Rep 1', axlabel='Bottleneck Radius (Angstroms)')
sns.distplot(df_6agf_2_pore, ax=axes[0], label='Rep 2')
sns.distplot(df_6agf_3_pore, ax=axes[0], label='Rep 3')
axes[0].set_title("Pore")

sns.distplot(df_6agf_1_sf, ax=axes[1], label='Rep 1', axlabel='Bottleneck Radius (Angstroms)')
sns.distplot(df_6agf_2_sf, ax=axes[1], label='Rep 2')
sns.distplot(df_6agf_3_sf, ax=axes[1], label='Rep 3')
axes[1].set_title("Selectivity Filter")

plt.legend()
```
