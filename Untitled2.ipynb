```python
def moving_avg(df):
    smoothed = np.convolve(df, np.ones(500), 'valid')/500
    smoothed_add = []
    i = 1
    while i < len(df)+1:
        if i < 251:
            smoothed_add.append(smoothed[0])
        elif i > 250 and i < len(df)-249:
            smoothed_add.append(smoothed[i-250])
        else: 
            smoothed_add.append(smoothed[-1])
        i += 1
            
    return smoothed_add
```
