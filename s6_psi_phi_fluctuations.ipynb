```python
mean_d1_phi = np.mean(d1_phi[1], axis=0)
mean_d1_psi = np.mean(d1_psi[1], axis=0)
mean_d2_phi = np.mean(d2_phi[1], axis=0)
mean_d2_psi = np.mean(d2_psi[1], axis=0)
mean_d3_phi = np.mean(d3_phi[1], axis=0)
mean_d3_psi = np.mean(d3_psi[1], axis=0)
mean_d4_phi = np.mean(d4_phi[1], axis=0)
mean_d4_psi = np.mean(d4_psi[1], axis=0)
```

```python
mean_d1_phi[14]
```

```python
import math

d1_devs = []
j = 0
while j < 27:
    res1 = []
    i = 0
    while i < 3000:
        phit1 = d1_phi[1][i,j]
        psit1 = d1_psi[1][i,j+1]
        dev1 = math.sqrt(((phit1-mean_d1_phi[1])**2) + ((psit1-mean_d1_psi[0])**2))
        res1.append(dev1)
        i +=1
    d1_devs.append(res1)
    j +=1

sns.heatmap(d1_devs, vmin=0, vmax=2.4, yticklabels=np.arange(255, 282), xticklabels=500)
    
```
